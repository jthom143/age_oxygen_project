```python
def solubility_CFC12(T, S):
    a1 = -124.4395
    a2 = 185.4299
    a3 = 51.6383
    b1 = -0.149779
    b2 = 0.094668
    b3 = -0.0160043
    
    T = T+273.15

    lnF = a1 + a2*(100/T) + a3*np.log(T/100)  + S*(b1 + b2*(T/100) + b3*((T/100)**2))
    F = np.exp(lnF)
    
    #shape_array = T.shape
    #F = np.ones(shape_array)*np.nan
    #for i in range(0, shape_array[0]):
    #    for j in range(0,shape_array[1]):
    #        F_ij = a1 + a2*(100/T[i,j]) + a3*np.log(T[i,j]/100)  + S[i,j]*(b1 + b2*(T[i,j]/100) + b3*(T[i,j]/100)**2)
    #        F[i,j] = np.exp(F_ij)
    return  F
```

### Load Data

```python
PATH = '/RESEARCH/chapter3/data/Darryn_age/'
age_1990  = iris.load_cube(PATH+'p18_dg10.nc', 'mean age 1') 
age_2000  = iris.load_cube(PATH+'p18_dg10.nc', 'mean age 2')
cfc_1990  = iris.load_cube(PATH+'p18_dg10.nc', 'cfc12 1')
cfc_2000  = iris.load_cube(PATH+'p18_dg10.nc', 'cfc12 2')
temp_1990 = iris.load_cube(PATH+'p18_dg10.nc', 'temperature 1')
temp_2000 = iris.load_cube(PATH+'p18_dg10.nc', 'temperature 2')
salt_1990 = iris.load_cube(PATH+'p18_dg10.nc', 'salinity 1')
salt_2000 = iris.load_cube(PATH+'p18_dg10.nc', 'salinity 2')
```

### Plot CFC Data

```python
F_1 = solubility_CFC12(temp_1990.data, salt_1990.data)
CFC_1990_plot = cfc_1990.data/F_1

F_2 = solubility_CFC12(temp_2000.data, salt_2000.data)
CFC_2000_plot = cfc_2000.data/F_2

sigma_1990 = gsw.sigma0(salt_1990.data, temp_1990.data)
sigma_2000 = gsw.sigma0(salt_2000.data, temp_2000.data)

lats = salt_1990.coord('latitude').points
depth = salt_1990.coord('depth').points
```

```python
plt.figure(figsize = (16,8))

clevs = np.arange(0, 600, 50)
clevs_diff = np.arange(-100, 105, 5)

ax1 = plt.subplot(2,2,1)
plt.contourf(lats, depth, CFC_1990_plot, clevs, cmap = 'YlOrRd', extend = 'max')
cb = plt.colorbar()
CS = plt.contour(lats, depth, sigma_1990, colors = 'k', levels=[26.6, 27.0, 27.2, 27.6])
plt.clabel(CS, fontsize=9, inline=1)
plt.ylim([0, 1500])
plt.xlim([-65, -15])
ax1.invert_yaxis()
cb.set_label('pCFC-12 [pptv]', fontsize = 14)
plt.title('Observed pCFC-12 1991')

ax2 = plt.subplot(2,2,2)
plt.contourf(lats, depth, CFC_2000_plot, clevs, cmap = 'YlOrRd', extend = 'max')
cb = plt.colorbar()
CS = plt.contour(lats, depth, sigma_2000, colors = 'k', levels=[26.6, 27.0, 27.2, 27.6])
plt.clabel(CS, fontsize=9, inline=1)
plt.ylim([0, 1500])
plt.xlim([-65, -15])
ax2.invert_yaxis()
cb.set_label('pCFC-12 [pptv]', fontsize = 14)
plt.title('Observed pCFC-12 2005')

ax3 = plt.subplot(2,2,3)
plt.contourf(lats, depth, CFC_2000_plot-CFC_1990_plot, clevs_diff, cmap = 'RdBu_r', extend = 'both')
cb = plt.colorbar()
CS = plt.contour(lats, depth, sigma_1990, colors = 'k', levels=[26.6, 27.0, 27.2, 27.6])
plt.clabel(CS, fontsize=9, inline=1)
plt.ylim([0, 1500])
plt.xlim([-65, -15])
ax3.invert_yaxis()
cb.set_label('pCFC-12 [pptv]', fontsize = 14)
plt.title('Observed pCFC-12 2005 - 1991')
```

```python
constraint = iris.Constraint(latitude=lambda x: x <= -50)
sigma0_1990 = gsw.sigma0(salt_1990.extract(constraint).data, temp_1990.extract(constraint).data)
sigma0_2000 = gsw.sigma0(salt_2000.extract(constraint).data, temp_2000.extract(constraint).data)

age_aabw = ma.masked_where(sigma0_1990 <= 27.2, age_1990.extract(constraint).data)
age_aabw_2d_1990 = ma.masked_where(sigma0_1990 >= 27.6, age_aabw)

age_aabw = ma.masked_where(sigma0_2000 <= 27.2, age_2000.extract(constraint).data)
age_aabw_2d_2000 = ma.masked_where(sigma0_2000 >= 27.6, age_aabw)
```

```python
constraint = iris.Constraint(latitude=lambda x: -50 < x < -20)
sigma0_1990_samw = gsw.sigma0(salt_1990.extract(constraint).data, temp_1990.extract(constraint).data)
sigma0_2000_samw = gsw.sigma0(salt_2000.extract(constraint).data, temp_2000.extract(constraint).data)

age_samw = ma.masked_where(sigma0_1990_samw <= 26.6, age_1990.extract(constraint).data)
age_samw_2d_1990 = ma.masked_where(sigma0_1990_samw >= 27.0, age_samw)

age_samw = ma.masked_where(sigma0_2000_samw <= 26.6, age_2000.extract(constraint).data)
age_samw_2d_2000 = ma.masked_where(sigma0_2000_samw >= 27.0, age_samw)
```

```python
clevs = np.arange(-90, 100, 10)

constraint = iris.Constraint(latitude=lambda x: x <= -50)
lats = salt_2000.extract(constraint).coord('latitude').points
plt.figure(figsize=(14,6))
ax = plt.gca()
plt.contourf(lats, depth, ((age_aabw_2d_2000 - age_aabw_2d_1990)/age_aabw_2d_1990)*100, clevs, cmap = 'RdBu_r', extend = 'both')
plt.colorbar()
CS = plt.contour(lats, depth, sigma0_1990, colors = 'k', levels=[26.6, 27.0, 27.2, 27.6])
plt.clabel(CS, fontsize=9, inline=1)
ax.invert_yaxis()
plt.xlim([-65, -15])

constraint = iris.Constraint(latitude=lambda x: -50 < x < -20)
lats = salt_2000.extract(constraint).coord('latitude').points
plt.contourf(lats, depth, ((age_samw_2d_2000 - age_samw_2d_1990)/age_samw_2d_1990)*100, clevs,cmap = 'RdBu_r', extend = 'both')
CS = plt.contour(lats, depth, sigma0_1990_samw, colors = 'k', levels=[26.6, 27.0, 27.2, 27.6])
plt.clabel(CS, fontsize=9, inline=1)
plt.xlim([-65, -15])
```
