###### Interpolate to find the depth of the 26 isopycnal

```python
def find_sigma_depth(sigma, sigma_level, depth_array, sigma_depth):
    for t in range(0,500):
        for y in range(0, len(lons)):
            for x in range(0, len(lats)):
                f = interp.interp1d(sigma[t,:,x,y], depth_array, bounds_error=False)
                sigma_depth[t,x,y] = f(sigma_level)
    return sigma_depth
```

```python
sigma_data = sigma0[0, :, :, 40].data
lons = sigma0.coord('longitude').points
lats = sigma0.coord('latitude').points
depth = sigma0.coord('tcell pstar').points

depth26 = np.ones((500,80,120)) * np.nan
 

%time
depth26 = find_sigma_depth(sigma0.data, 26.0, depth, depth26)
```

```python
plt.figure(figsize=(10,8))
plt.ylim([0, 500])
qplt.contourf(sigma0[0, :, :, 40], 20, cmap = 'jet')
plt.plot(lats, depth26[0,:,40])
CS = plt.contour(lats, depth, sigma0[0, :, :, 40].data, colors='k', levels=[26])
plt.clabel(CS, fontsize=9, inline=1)
```

```python
plt.figure(figsize=(12,10))
ax = plt.axes(projection=ccrs.PlateCarree())
clevs = np.arange(0, 650, 50)
plt.contourf(lons, lats, depth26[0,:,:], clevs, cmap = cmaps.viridis, extend = 'max')
cb = plt.colorbar(orientation='horizontal')
cb.set_label('Depth [dbars]', fontsize = 14)
plt.title('Depth of $\sigma_{\Theta}$=26.0', fontsize = 14)
ax.coastlines(resolution='50m', color='black', lw = 1.5)
ax.add_feature(cfeature.LAND, facecolor='0.75')

plt.savefig('sigma_26_depth.pdf', bbox_inches='tight')
```

```python
def var_on_isopycnal(var, depth, sigma_depth, var_isopycnal):
    for t in range(0, 500):
        for y in range(0, len(lons)):
            for x in range(0, len(lats)):
                f = interp.interp1d(depth, var[t,:,x,y], bounds_error=False)
                var_isopycnal[t,x,y] = f(sigma_depth[t,x,y])
    return var_isopycnal
```

```python
age26 = np.ones((500, 80,120)) * np.nan
var = age.data

age26 = var_on_isopycnal(var, depth, depth26, age26)
```

```python
plt.figure(figsize=(12,10))
ax = plt.axes(projection=ccrs.PlateCarree())
clevs = np.arange(0, 320, 20)
plt.contourf(lons, lats, age26[0,:,:], clevs, cmap = cmaps.viridis, extend = 'max')
cb = plt.colorbar(orientation='horizontal')
cb.set_label('Age [years]', fontsize = 14)
plt.title('Age on $\sigma_{\Theta}$=26.0', fontsize = 14)
ax.coastlines(resolution='50m', color='black', lw = 1.5)
ax.add_feature(cfeature.LAND, facecolor='0.75')

plt.savefig('sigma_26_age.pdf', bbox_inches='tight')
```

```python
o226 = np.ones((500, 80,120)) * np.nan
var = o2.data

o226 = var_on_isopycnal(var, depth, depth26, o226)
```

```python
plt.figure(figsize=(12,10))
ax = plt.axes(projection=ccrs.PlateCarree())
clevs = np.arange(0, 320, 20)
plt.contourf(lons, lats, o226[0,:,:]*1e6, clevs, cmap = cmaps.viridis, extend = 'max')
cb = plt.colorbar(orientation='horizontal')
cb.set_label('Dissoved Oxygen [umol/kg]', fontsize = 14)
plt.title('O$_2$ on $\sigma_{\Theta}$=26.0', fontsize = 14)
ax.coastlines(resolution='50m', color='black', lw = 1.5)
ax.add_feature(cfeature.LAND, facecolor='0.75')

plt.savefig('sigma_26_o2.pdf', bbox_inches='tight')
```

```python
o2_26_cube = iris.cube.Cube(o226,
                      dim_coords_and_dims=[(o2.coord('time'), 0), (o2.coord('latitude'), 1), (o2.coord('longitude'), 2)])

age_26_cube = iris.cube.Cube(age26,
                      dim_coords_and_dims=[(o2.coord('time'), 0), (o2.coord('latitude'), 1), (o2.coord('longitude'), 2)])
```

```python
import iris.analysis.stats
correlation = iris.analysis.stats.pearsonr(age_26_cube, o2_26_cube, corr_coords='time')
```

```python
plt.figure(figsize=(12,10))
ax = plt.axes(projection=ccrs.PlateCarree())
clevs = np.arange(0, 320, 20)
plt.contourf(lons, lats, correlation.data, cmap = 'RdBu_r')
plt.colorbar(orientation='horizontal')
plt.title('Age-O$_2$ Correlation on $\sigma_{\Theta}$=26.0', fontsize = 14)
ax.coastlines(resolution='50m', color='black', lw = 1.5)
ax.add_feature(cfeature.LAND, facecolor='0.75')
plt.plot([-150, -150],[-60, 50], color = 'k')

plt.savefig('sigma_26_correlation.pdf', bbox_inches='tight')
```
