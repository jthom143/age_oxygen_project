###### Load Data

```python
PATH = '/RESEARCH/chapter3/data/newCO2_control_800/'
o2 = iris.load_cube(PATH+'o2.nc')
o2 = o2[-500:]

age = iris.load_cube(PATH+'residency_age_surface.nc')
age.coord('Time').rename('time')

neutral_rho = iris.load_cube(PATH+'neutral_rho.nc')
neutral_rho = neutral_rho - 1000
```

###### Interpolate to find the depth of the 27.0 neutral density surface

```python
def find_sigma_depth(sigma, sigma_level, depth_array, sigma_depth):
    for t in range(0,500):
        for y in range(0, len(lons)):
            for x in range(0, len(lats)):
                f = interp.interp1d(sigma[t,:,x,y], depth_array, bounds_error=False)
                sigma_depth[t,x,y] = f(sigma_level)
    return sigma_depth
```

```python
def var_on_isopycnal(var, depth, sigma_depth, var_isopycnal):
    for t in range(0, 500):
        for y in range(0, len(lons)):
            for x in range(0, len(lats)):
                f = interp.interp1d(depth, var[t,:,x,y], bounds_error=False)
                var_isopycnal[t,x,y] = f(sigma_depth[t,x,y])
    return var_isopycnal
```

```python
lons = neutral_rho.coord('longitude').points
lats = neutral_rho.coord('latitude').points
depth = neutral_rho.coord('tcell pstar').points

depth27 = np.ones((500,80,120)) * np.nan
 
depth27 = find_sigma_depth(neutral_rho.data, 26.6, depth, depth27)
```

```python
plt.figure(figsize=(12,10))
ax = plt.axes(projection=ccrs.PlateCarree())
clevs = np.arange(0, 1050, 50)
plt.contourf(lons, lats, depth27[0,:,:], clevs, cmap = cmaps.viridis, extend = 'max')
cb = plt.colorbar(orientation='horizontal')
cb.set_label('Depth [dbars]', fontsize = 14)
plt.title('Depth of $\gamma{n}$=26.6', fontsize = 14)
ax.coastlines(resolution='50m', color='black', lw = 1.5)
ax.add_feature(cfeature.LAND, facecolor='0.75')
```

```python
age27 = np.ones((500,80,120)) * np.nan
var = age.data
age27 = var_on_isopycnal(var, depth, depth27, age27)

o227 = np.ones((500,80,120)) * np.nan
var = o2.data
o227 = var_on_isopycnal(var, depth, depth27, o227)
```

```python
plt.figure(figsize=(12,10))
ax = plt.axes(projection=ccrs.PlateCarree())
clevs = np.arange(0, 800, 20)
plt.contourf(lons, lats, age27[0,:,:], clevs, cmap = cmaps.viridis, extend = 'max')
cb = plt.colorbar(orientation='horizontal')
cb.set_label('Age [years]', fontsize = 14)
plt.title('Age on $\gamma_{n}$=26.6', fontsize = 14)
ax.coastlines(resolution='50m', color='black', lw = 1.5)
ax.add_feature(cfeature.LAND, facecolor='0.75')
```

```python
plt.figure(figsize=(12,10))
ax = plt.axes(projection=ccrs.PlateCarree())
clevs = np.arange(0, 320, 20)
plt.contourf(lons, lats, o227[0,:,:]*1e6,clevs, cmap = cmaps.viridis, extend = 'max')
cb = plt.colorbar(orientation='horizontal')
cb.set_label('Dissoved Oxygen [umol/kg]', fontsize = 14)
plt.title('O2 on $\gamma_{n}$=26.6', fontsize = 14)
ax.coastlines(resolution='50m', color='black', lw = 1.5)
ax.add_feature(cfeature.LAND, facecolor='0.75')
```

###### Calculate Correlation

```python
# Change into cubes 
o2_27 = iris.cube.Cube(o227, long_name='oxygen', units=o2.units, 
                            dim_coords_and_dims=[(o2.coord('time'), 0), 
                                                 (o2.coord('latitude'), 1),
                                                 (o2.coord('longitude'), 2)])
age_27 = iris.cube.Cube(age27,  long_name='age', units=age.units,  
                                dim_coords_and_dims=[(age.coord('time'), 0), 
                                                      (age.coord('latitude'), 1),
                                                      (age.coord('longitude'), 2)])
```

```python
corr = iris.analysis.stats.pearsonr(o2_27, age_27, corr_coords=['time'])
```
