### Function to Regrid onto Smaller Grid

```python
def SAMW_smaller_grid(var_cube, z_grid, x_grid):
    
    # Dimensions of original cube
    time = var_cube.coord('time').points
    lats = var_cube.coord('latitude').points
    depths = var_cube.coord('tcell pstar').points
    
    new_grid = np.ones((len(time), len(z_grid), len(x_grid)))*np.nan
    
    for t in range(0, len(time)):
        old_grid_data = var_cube[t,:,:].data

        #create mesh
        X, Y = np.meshgrid(lats, depths)
        XI, YI = np.meshgrid(x_grid, z_grid)
        
        #interp
        new = griddata((X.flatten(),Y.flatten()), old_grid_data.flatten(), (XI,YI),method='linear')
        
        new_grid[t,:,:] = new
            
    return new_grid
```

### Calculate Age and Oxygen for CDW and SAMW

Define CDW and SAMW water masses following (approximately) same method used in Waugh et al., 2013: 

- SAMW: Water that lies on potential density surfaces between 26.6 - 27.0 and between latitudes 20S - 50S 
- CDW:  Water that lies on potential density surfaces between 27.2 - 27.7 and South of 50S. 

###### Mask grid points outside of density threshold

```python
lat_constraint = iris.Constraint(latitude=lambda y: -50 < y <= -20)
age_samw_region = age.extract(lat_constraint)
o2_samw_region  = o2.extract(lat_constraint)
sigma0_samw_region = sigma0.extract(lat_constraint)

age_samw = ma.masked_where(sigma0_samw_region.data >= 27.0, age_samw_region.data)
age_samw = ma.masked_where(sigma0_samw_region.data <= 26.6, age_samw)

o2_samw = ma.masked_where(sigma0_samw_region.data >= 27.0, o2_samw_region.data)
o2_samw = ma.masked_where(sigma0_samw_region.data <= 26.6, o2_samw)
```

```python
slope, intercept, r_value, p_value, std_err = stats.mstats.linregress(age_samw.flatten(),o2_samw.flatten()*1e5)
print slope
print intercept
print r_value**2
```

```python
plt.scatter(age_samw, o2_samw*1e5)
plt.ylabel('Oxygen [umol/kg]', fontsize = 14)
plt.xlabel('Age [years]', fontsize = 14)
plt.title('SAMW Age- Oxygen Relationship', fontsize = 14)
x = np.arange(-100, 701, 1)
plt.plot(x, (slope*x)+intercept, color = 'k', lw = 2)
plt.xlim([-100, 700])
```

```python
lat_constraint = iris.Constraint(latitude=lambda y: y <= -50)
age_cdw_region = age.extract(lat_constraint)
o2_cdw_region  = o2.extract(lat_constraint)
sigma0_cdw_region = sigma0.extract(lat_constraint)

age_cdw = ma.masked_where(sigma0_cdw_region.data >= 27.7, age_cdw_region.data)
age_cdw = ma.masked_where(sigma0_cdw_region.data <= 27.2, age_cdw)

o2_cdw = ma.masked_where(sigma0_cdw_region.data >= 27.7, o2_cdw_region.data)
o2_cdw = ma.masked_where(sigma0_cdw_region.data <= 27.2, o2_cdw)
```

```python
slope, intercept, r_value, p_value, std_err = stats.mstats.linregress(age_cdw.flatten(),o2_cdw.flatten()*1e6)
print slope
print intercept
print r_value**2
```

```python
plt.scatter(age_cdw, o2_cdw*1e6)
plt.ylabel('Oxygen [umol/kg]', fontsize = 14)
plt.xlabel('Age [years]', fontsize = 14)
plt.title('CDW Age - Oxygen Relationship', fontsize = 14)
x = np.arange(-100, 801, 1)
plt.plot(x, (slope*x)+intercept, color = 'k', lw = 2)
plt.xlim([-100, 800])
```
