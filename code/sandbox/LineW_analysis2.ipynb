###### Isolate Line W and Reorder observations to ascending latitude and depth

```python
linew = data[data.cruise==324]
linew = linew.sort(['latitude', 'depth'])
```

###### Determine Latitude and Longitude of Each Sample Point

```python
a = linew.latitude.values
indexes = np.unique(a, return_index=True)[1]
linew_lats = [a[index] for index in sorted(indexes)]

b = linew.longitude.values
indexes = np.unique(b, return_index=True)[1]
linew_lons = [b[index] for index in sorted(indexes)]
```

###### Plot Line W

```python
# Plot Transects
import matplotlib.pyplot as plt
import cartopy.crs as ccrs

plt.figure(figsize=(12,10))
ax = plt.axes(projection=ccrs.PlateCarree())
ax.set_extent((-75, -65, 35, 45), crs=ccrs.PlateCarree())
plt.plot(linew_lons, linew_lats, marker = '*', transform=ccrs.PlateCarree())

ax.stock_img()
ax.coastlines()
```

```python
grid_z, grid_x = np.mgrid[0:5000:51j, 37:41:17j]

points = np.column_stack((linew.depth.values, linew.latitude.values))
values = linew.oxygen.values
b = griddata(points, linew.oxygen.values, (grid_z, grid_x), method = 'linear')
```

```python
plt.figure(figsize=(14,8))
clevs = np.arange(150, 300.5, 0.5)
ax = plt.gca()
plt.contourf(grid_x, grid_z, b, clevs, cmap = cmaps.viridis, extend = 'both')
plt.plot(linew.latitude, linew.depth, ls = '', marker = '.', color = 'grey')
ax.invert_yaxis()
cb = plt.colorbar()
cb.set_label('umol/kg', fontsize = 14)
plt.ylabel('Depth [m]', fontsize = 14)
plt.xlabel('Latitude', fontsize = 14)
plt.title('GLODAPv2 Line W - Oxygen', fontsize = 14)
plt.xlim([37.5, 40.25])
```

###### Calculate Correatlion

```python
import iris.analysis.stats
```

```python
o2_model_linew = np.ones((len(o2.coord('time').points), len(o2.coord('tcell pstar').points), len(lat_index)))* np.nan
age_model_linew = np.ones((len(o2.coord('time').points), len(o2.coord('tcell pstar').points), len(lat_index)))* np.nan
for t in range(len(o2.coord('time').points)):
    for k in range(len(o2.coord('tcell pstar').points)):
        for i in range(len(lat_index)):
            o2_model_linew[t,k,i] = o2_regrid[t, k, lat_index[i], lon_index[i]].data
            age_model_linew[t,k,i] = age_regrid[t, k, lat_index[i], lon_index[i]].data
```

```python
latitude = iris.coords.DimCoord(linew_lats, standard_name='latitude', units='degrees')
cube_o2 = iris.cube.Cube(o2_model_linew, standard_name = 'moles_of_oxygen_per_unit_mass_in_sea_water',  
                                long_name = 'modeled oxygen interpolated onto lineW track', units = 'mol/kg', 
                                dim_coords_and_dims=[(o2.coord('time'), 0), (o2.coord('tcell pstar'), 1),(latitude, 2)])
```

```python
cube_age = iris.cube.Cube(age_model_linew, standard_name = 'sea_water_age_since_surface_contact',  
                                 long_name = 'modeled age interpolated onto lineW track', units = 'yr', 
                                 dim_coords_and_dims=[(o2.coord('time'), 0), (o2.coord('tcell pstar'), 1),(latitude, 2)])
```

```python
correlation = iris.analysis.stats.pearsonr(cube_age, cube_o2, corr_coords='time') 
```

```python
plt.figure(figsize=(14,8))
clevs = np.arange(-1, 1.1, 0.1)
ax = plt.gca()
plt.contourf(linew_lats, o2.coord('tcell pstar').points, correlation.data, clevs, cmap = 'RdBu_r')
ax.invert_yaxis()
cb = plt.colorbar()
plt.ylabel('Depth [m]', fontsize = 14)
plt.xlabel('Latitude', fontsize = 14)
plt.title('Model Line W - Age-Oxygen Correlation', fontsize = 14)
```
