###### Load Packages

```python
import numpy as np
import pandas as pd
import sys
import iris.quickplot as qplt
import matplotlib.pyplot as plt
from scipy.interpolate import griddata
sys.path.append('/RESEARCH/paper_ocean_heat_carbon/code/python')
import colormaps as cmaps
```

```python
%matplotlib inline
```

###### Load Data

```python
fname ='6_33RR20050106_hy1.csv'
path = '/RESEARCH/chapter3/data/GO_SHIP/P16/33RR200501/'
data = pd.read_csv(path+fname, header = 78, na_values='-999.000')
```

```python
print data.columns
```

```python
CFC_data = data[['DATE', 'LATITUDE', 'LONGITUDE', 'CTDPRS', 'CFC-12', 'CFC-12_FLAG_W']].copy()
CFC_data.columns = ['date', 'latitude', 'longitude', 'press', 'CFC_12', 'CFC_12_FLAG_W']

print CFC_data.head()
```

```python
CFC_data=CFC_data.drop(data.index[[0]])
CFC_data = CFC_data[CFC_data.CFC_12_FLAG_W!=9]
```

###### Determine Latitude and Longitude of Each Sample Point

```python
lats = CFC_data.latitude.values
lons = CFC_data.longitude.values
```

```python
# Plot Transects
import matplotlib.pyplot as plt
import cartopy.crs as ccrs

plt.figure(figsize=(12,10))
ax = plt.axes(projection=ccrs.PlateCarree())
ax.set_extent((-180, 180, -90, 0), crs=ccrs.PlateCarree())
plt.plot(lons, lats, marker = '*',ms = 2, transform=ccrs.PlateCarree())

ax.stock_img()
ax.coastlines()
```

```python
depi = CFC_data.press.values
lati = CFC_data.latitude.values

lat = np.arange(-72.,-16.,1)
depth = np.arange(0,5000,250)
```

```python
#create mesh
XI, YI = np.meshgrid(lat,depth)
```

```python
new_grid=griddata((lati.flatten(),depi.flatten()), CFC_data.CFC_12.values.flatten(), (XI,YI), method='linear')
```

```python
plt.figure(figsize=(14,8))
ax = plt.gca()
clevs = np.arange(0, 2.4, 0.2)
plt.contourf(XI, YI, new_grid, clevs, cmap = cmaps.viridis, extend = 'both')
cb = plt.colorbar()
ax.invert_yaxis()
plt.ylabel('Depth [dbars]', fontsize = 14)
cb.set_label('Pmol/kg', fontsize = 14)
plt.xlabel('Latitude', fontsize = 14)
plt.title('Line P16S - CFC Concentration', fontsize = 14)
```
