```python
def solubility_CFC12(T, S):
    a1 = -124.4395
    a2 = 185.4299
    a3 = 51.6383
    b1 = -0.149779
    b2 = 0.094668
    b3 = -0.0160043
    
    T = T+273.15

    lnF = a1 + a2*(100/T) + a3*np.log(T/100)  + S*(b1 + b2*(T/100) + b3*((T/100)**2))
    F = np.exp(lnF)
    
    #shape_array = T.shape
    #F = np.ones(shape_array)*np.nan
    #for i in range(0, shape_array[0]):
    #    for j in range(0,shape_array[1]):
    #        F_ij = a1 + a2*(100/T[i,j]) + a3*np.log(T[i,j]/100)  + S[i,j]*(b1 + b2*(T[i,j]/100) + b3*(T[i,j]/100)**2)
    #        F[i,j] = np.exp(F_ij)
    return  F
```

### Watermass Analysis - Recreate Figure 3 from Waugh et al, 2013

```python
age_samw_1990 = {}
age_samw_2000 = {}

for key, cube in data.iteritems():
    # Constrain Data: 
    constraint = iris.Constraint(latitude=lambda x: -50 < x < -20)
    sigma0_1990 = gsw.sigma0(cube['salt_1990'].extract(constraint).data, cube['temp_1990'].extract(constraint).data)
    sigma0_2000 = gsw.sigma0(cube['salt_2000'].extract(constraint).data, cube['temp_2000'].extract(constraint).data)
    
    # 1990s:
    age_samw = ma.masked_where(sigma0_1990 <= 26.6, cube['age_1990'].extract(constraint).data)
    age_samw_2d_1990 = ma.masked_where(sigma0_1990 >= 27.0, age_samw)
    age_samw_2d_1990[age_samw_2d_1990 < -1e12] = np.nan # When data is less than -1e9 set to NaN
    age_samw_1990[key] = np.nanmean(age_samw_2d_1990)

    age_samw = ma.masked_where(sigma0_2000 <= 26.6, cube['age_2000'].extract(constraint).data )
    age_samw_2d_2000 = ma.masked_where(sigma0_2000 >= 27.0, age_samw)
    age_samw_2d_2000[age_samw_2d_2000 < -1e12] = np.nan
    age_samw_2000[key] = np.nanmean(age_samw_2d_2000)
```

```python
age_aabw_1990 = {}
age_aabw_2000 = {}

for key, cube in data.iteritems():
    # Constrain Data: 
    constraint = iris.Constraint(latitude=lambda x: x < -50)
    sigma0_1990 = gsw.sigma0(cube['salt_1990'].extract(constraint).data, cube['temp_1990'].extract(constraint).data)
    sigma0_2000 = gsw.sigma0(cube['salt_2000'].extract(constraint).data, cube['temp_2000'].extract(constraint).data)
    
    # 1990s:
    age_aabw = ma.masked_where(sigma0_1990 <= 27.2, cube['age_1990'].extract(constraint).data)
    age_aabw_2d_1990 = ma.masked_where(sigma0_1990 >= 27.6, age_aabw)
    age_aabw_2d_1990[age_aabw_2d_1990 < -1e9] = np.nan # When data is less than -1e9 set to NaN
    age_aabw_1990[key] = np.nanmean(age_aabw_2d_1990)

    age_aabw = ma.masked_where(sigma0_2000 <= 27.2, cube['age_2000'].extract(constraint).data )
    age_aabw_2d_2000 = ma.masked_where(sigma0_2000 >= 27.6, age_aabw)
    age_aabw_2d_2000[age_aabw_2d_2000 < -1e9] = np.nan
    age_aabw_2000[key] = np.nanmean(age_aabw_2d_2000)
```
