### Define Functions

###### Solubility of CFC-12:

Empirical formula for solubility of CFC-12 from Warner & Weiss (1985)

```python
def solubility_CFC12(T, S):
    a1 = -124.4395
    a2 = 185.4299
    a3 = 51.6383
    b1 = -0.149779
    b2 = 0.094668
    b3 = -0.0160043
    
    T = T+273.15

    lnF = a1 + a2*(100/T) + a3*np.log(T/100)  + S*(b1 + b2*(T/100) + b3*((T/100)**2))
    F = np.exp(lnF)
    
    #shape_array = T.shape
    #F = np.ones(shape_array)*np.nan
    #for i in range(0, shape_array[0]):
    #    for j in range(0,shape_array[1]):
    #        F_ij = a1 + a2*(100/T[i,j]) + a3*np.log(T[i,j]/100)  + S[i,j]*(b1 + b2*(T[i,j]/100) + b3*(T[i,j]/100)**2)
    #        F[i,j] = np.exp(F_ij)
    return  F

```

### Regrid Each Hydrography Track onto Regular Grid (1 deg lat x 250 m)

###### 1991 Cruise:

```python
# Old Depth and Latitude Coordinates
depi = CFC_data_91.press.values
lati = CFC_data_91.latitude.values

# New Grid 
lat = np.arange(-72.,-16.,1)
depth = np.arange(0,5000,250)

XI, YI = np.meshgrid(lat,depth)

# Regrid
CFC_data_91_grid = griddata((lati.flatten(),depi.flatten()), CFC_data_91.CFC_12.values.flatten(), (XI,YI), method='linear')
temp_data_91_grid = griddata((lati.flatten(),depi.flatten()), CFC_data_91.temp.values.flatten(), (XI,YI), method='linear')
salt_data_91_grid = griddata((lati.flatten(),depi.flatten()), CFC_data_91.salt.values.flatten(), (XI,YI), method='linear')
```

###### 2005 Cruise:

```python
# Old Depth and Latitude Coordinates
depi = CFC_data_05.press.values
lati = CFC_data_05.latitude.values

# New Grid 
lat = np.arange(-72.,-16.,1)
depth = np.arange(0,5000,250)

XI, YI = np.meshgrid(lat,depth)

# Regrid
CFC_data_05_grid = griddata((lati.flatten(),depi.flatten()), CFC_data_05.CFC_12.values.flatten(), (XI,YI), method='linear')
temp_data_05_grid = griddata((lati.flatten(),depi.flatten()), CFC_data_05.temp.values.flatten(), (XI,YI), method='linear')
salt_data_05_grid = griddata((lati.flatten(),depi.flatten()), CFC_data_05.salt.values.flatten(), (XI,YI), method='linear')
```

### Create Figure of 1991 Observations

```python
F_91 = solubility_CFC12(temp_data_91_grid, salt_data_91_grid)
CFC_data_91_grid_pptv = CFC_data_91_grid/F_91
```

```python
sigma0_91 = gsw.sigma0(salt_data_91_grid, temp_data_91_grid)

plt.figure(figsize=(14,6))
ax = plt.gca()
clevs = np.arange(0, 575, 25)
plt.contourf(XI, YI, CFC_data_91_grid_pptv , clevs, cmap = 'YlOrRd', extend = 'max')
cb = plt.colorbar()
CS = plt.contour(XI, YI, sigma0_91, colors = 'k', levels=[26.6, 27.0, 27.2, 27.6])
plt.clabel(CS, fontsize=9, inline=1)
plt.ylim([200, 1700])
plt.xlim([-65, -15])
ax.invert_yaxis()
plt.ylabel('Depth [dbars]', fontsize = 14)
cb.set_label('pCFC-12 [pptv]', fontsize = 14)
plt.xlabel('Latitude', fontsize = 14)
plt.title('P16S 1991 Observations', fontsize = 14)
```
