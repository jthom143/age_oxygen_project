###### Load Packages

```python
import iris
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
import iris.analysis.stats
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import sys
sys.path.append('/RESEARCH/paper_ocean_heat_carbon/code/python')
import colormaps as cmaps
```

```python
%matplotlib inline
```

###### Load Data

```python
PATH = '/RESEARCH/chapter3/data/newCO2_control_800/'
o2 = iris.load_cube(PATH+'o2.nc')
o2 = o2[-500:]

age = iris.load_cube(PATH+'residency_age_surface.nc')
age.coord('Time').rename('time')

sigma0 = iris.load_cube(PATH+'/derived/sigma0.nc')
```

###### Map of Various Longitude Lines

```python
plt.figure(figsize=(12,10))
ax = plt.axes(projection=ccrs.PlateCarree())
clevs = np.arange(0, 320, 20)
plt.title('Longitude Lines', fontsize = 14)
ax.stock_img()
plt.plot([-150, -150],[-60, 50], color = 'k')
plt.plot([-110, -110],[-60, 25], color = 'k')
plt.plot([-25, -25],[-60, 60], color = 'k')
plt.plot([0, 0],[-60, 5], color = 'k')
plt.plot([90, 90],[-60, 20], color = 'k')
plt.savefig('longitude_lines_map.png', bbox_inches='tight')
```

### Isolate Longitude line 150W to Match Repeat Hydrography Line P16

```python
lon_constraint = iris.Constraint(longitude=lambda y: -152 < y < -149)
lat_constraint = iris.Constraint(latitude=lambda x: -90 < x < 50)
o2_P16 = o2.extract(lon_constraint)
age_P16 = age.extract(lon_constraint)
sigma0_P16 = sigma0.extract(lon_constraint)

o2_P16 = o2_P16.extract(lat_constraint)
age_P16 = age_P16.extract(lat_constraint)
sigma0_P16 = sigma0_P16.extract(lat_constraint)
```

###### Calculate Climatology to make Age and Oxygen Maps

```python
o2_P16_clim = o2_P16.collapsed('time', iris.analysis.MEAN)
age_P16_clim = age_P16.collapsed('time', iris.analysis.MEAN)
sigma0_P16_clim = sigma0_P16.collapsed('time', iris.analysis.MEAN)
```

###### Plot Climatology for Age and Oxygen along ~Line P16

```python
lats = o2_P16.coord('latitude').points
depth = o2_P16.coord('tcell pstar').points
```

```python
plt.figure(figsize = (10,5))
ax = plt.gca()
plt.contourf(lats, depth, o2_P16_clim.data*1e6, cmap = cmaps.viridis)
cb = plt.colorbar()
CS = plt.contour(lats, depth, sigma0_P16_clim.data, colors='k', levels=[26, 27.2])
plt.clabel(CS, fontsize=9, inline=1)
ax.invert_yaxis()
cb.set_label('umol/kg', fontsize = 14)
plt.ylabel('Depth [m]', fontsize = 14)
plt.xlabel('Latitude', fontsize = 14)
plt.title('Model Line P16 - Oxygen', fontsize = 14)
```

```python
plt.figure(figsize = (10,5))
ax = plt.gca()
plt.contourf(lats, depth, age_P16_clim.data, cmap = cmaps.viridis)
cb = plt.colorbar()
CS = plt.contour(lats, depth, sigma0_P16_clim.data, colors='k', levels=[26, 27.2])
plt.clabel(CS, fontsize=9, inline=1)
ax.invert_yaxis()
cb.set_label('yrs', fontsize = 14)
plt.ylabel('Depth [m]', fontsize = 14)
plt.xlabel('Latitude', fontsize = 14)
plt.title('Model Line P16 - Age', fontsize = 14)
```

###### Calculate Correlation

```python
correlation = iris.analysis.stats.pearsonr(age_P16, o2_P16, corr_coords='time') 
```

```python
plt.figure(figsize = (10,5))
ax = plt.gca()
clevs = np.arange(-1, 1.25, 0.25)
plt.contourf(lats, depth, correlation.data, clevs, cmap = 'RdBu_r')
cb = plt.colorbar()
CS = plt.contour(lats, depth, sigma0_P16_clim.data, colors='k', levels=[26, 27.2])
plt.clabel(CS, fontsize=9, inline=1)
ax.invert_yaxis()
cb.set_label('Pearson Correlation Coefficient', fontsize = 14)
plt.ylabel('Depth [m]', fontsize = 14)
plt.xlabel('Latitude', fontsize = 14)
plt.title('Model Line P16 - Oxygen-Age Correlation', fontsize = 14)
```

```python
plt.figure(figsize = (10,3))
ax = plt.gca()
clevs = np.arange(-1, 1.25, 0.25)
plt.contourf(lats, depth, correlation.data, clevs, cmap = 'RdBu_r')
cb = plt.colorbar()
CS = plt.contour(lats, depth, sigma0_P16_clim.data, colors='k', levels=[26, 27.2])
plt.clabel(CS, fontsize=9, inline=1)
plt.ylim([0, 2000])
ax.invert_yaxis()
cb.set_label('Pearson Correlation Coefficient', fontsize = 14)
plt.ylabel('Depth [m]', fontsize = 14)
plt.xlabel('Latitude', fontsize = 14)
plt.title('Model Line P16 - Oxygen-Age Correlation', fontsize = 14)
```
