### Define Functions

```python
## Define Functions used in Analysis
def load_data(fname, path, header_no, variables, new_names, lon_lims):
    # Load Data
    data = pd.read_csv(path+fname, header = header_no, na_values='-999.000')
    
    # Isolate Desired Variables
    var = data[variables].copy()
    
    # Rename variables if desired
    var.columns = new_names
    
    # Drop data that is flagged
    var = var[var.oxygen_flag!=5]
    var = var[var.oxygen_flag!=9]

    var = var[var.po4_flag!=9]

    
    # Drop row with units
    var = var.drop(data.index[[0]])


    var = var[var.longitude<=lon_lims[0]]
    var = var[var.longitude>=lon_lims[1]] 
        
    return var
```

```python
def retrieve_old_grid(frame):
    depi = frame.press.values
    lati = frame.latitude.values
    old_grid = (lati.flatten(), depi.flatten())
    
    return old_grid
```

```python
def regrid(frame, old_grid, new_grid):
    o2_grid   = griddata(old_grid, frame.oxygen.values.flatten(), new_grid, 
                          method='linear')
    
    temp_grid   = griddata(old_grid, frame.temp.values.flatten(), new_grid, 
                           method='linear')
    
    salt_grid   = griddata(old_grid, frame.salt.values.flatten(), new_grid, 
                           method='linear')
    return o2_grid, temp_grid, salt_grid
```

### Load Data - Transect A16

```python
## Create a dictionary of raw data information:
data_dict = {'I08_1994': {'fname': '7_i08s_hy1.csv', 
                          'path': '/RESEARCH/chapter3/data/GO_SHIP/I08/316N145_5/', 'head_no':12},
             'I08_2005': {'fname': '6_33RR20070204_hy1.csv',
                          'path': '/RESEARCH/chapter3/data/GO_SHIP/I08/33RR20070204/', 'head_no': 36}
             }

```

```python
# Load Data
data = {}
for key, info in data_dict.iteritems():  
    if key == 'I08_1994' or key =='I08_2005' or key=='I08_2016':
        YY = [100, 80]
    
    data[key] = load_data(info['fname'], info['path'], info['head_no'],
                          ['DATE', 'LATITUDE', 'LONGITUDE', 'CTDPRS', 'OXYGEN', 'OXYGEN_FLAG_W', 
                           'CTDTMP', 'CTDSAL', 'PHSPHT',  'PHSPHT_FLAG_W'], 
                          new_names=['date', 'latitude', 'longitude', 'press', 'oxygen', 'oxygen_flag', 
                                     'temp', 'salt', 'po4', 'po4_flag'], 
                          lon_lims=YY)
```

### Regrid Each Hydrography Track onto Regular Grid (1 deg lat x 100 m)

```python
# New Grid 
lat = np.arange(-72.,-16.,1)
depth = np.arange(0,5000,100)   
XI, YI = np.meshgrid(lat,depth)
new_grid = (XI, YI)

o2_regrid = {}
sigma = {}
o2_aou = {}
for key, frame in data.iteritems():
    old_grid = retrieve_old_grid(frame)
    o2_regrid[key], temp, salt = regrid(frame, old_grid, new_grid)
    sigma[key] = gsw.sigma0(salt, temp)
```

# Figures of O2 Observations

```python
plt.figure(figsize=(14,8))
ax = plt.subplot(2,2,1)
clevs = np.arange(100,300,20)
plt.contourf(XI, YI, o2_regrid['I08_1994'], clevs,cmap = cmaps.viridis, extend = 'max')
plt.colorbar()

CS = plt.contour(XI, YI, sigma['I08_1994'], colors = 'k', levels=[26.6, 27.0, 27.2, 27.6])
plt.clabel(CS, fontsize=9, inline=1)
plt.ylim([200, 1700])
plt.xlim([-65, -15])
ax.invert_yaxis()
plt.ylabel('Depth [dbars]', fontsize = 14)
plt.title('A16S 1990s Observations', fontsize = 16)


ax = plt.subplot(2,2,2)
clevs = np.arange(100,300,20)
plt.contourf(XI, YI, o2_regrid['I08_2005'] , clevs,cmap = cmaps.viridis, extend = 'both')
plt.colorbar()

CS = plt.contour(XI, YI, sigma['I08_2005'], colors = 'k', levels=[26.6, 27.0, 27.2, 27.6])
plt.clabel(CS, fontsize=9, inline=1)
plt.ylim([200, 1700])
plt.xlim([-65, -15])
ax.invert_yaxis()
plt.ylabel('Depth [dbars]', fontsize = 14)
plt.title('A16S 2005 Observations', fontsize = 16)

ax = plt.subplot(2,2,3)
clevs = np.arange(-30,35,5)
plt.contourf(XI, YI, o2_regrid['I08_2005']-o2_regrid['I08_1994'] , clevs,cmap = 'RdBu_r', extend = 'both')
plt.colorbar()
CS = plt.contour(XI, YI, sigma['I08_1994'], colors = 'k', levels=[26.6, 27.0, 27.2, 27.6])
plt.clabel(CS, fontsize=9, inline=1)
plt.ylim([200, 1700])
plt.xlim([-65, -15])
ax.invert_yaxis()
plt.ylabel('Depth [dbars]', fontsize = 14)
plt.xlabel('Latitude', fontsize = 14)
plt.title('A16S 2005 - 1989 Observations', fontsize = 16)
```

# Calculate Precent Change in Each Watermass

### Subantarcitc Mode Water

```python
# New Grid 
lat = np.arange(-50.,-21.,1)
depth = np.arange(0,1500,100)   
XI_samw, YI_samw = np.meshgrid(lat,depth)
new_grid = (XI_samw, YI_samw)

o2_samw_2d = {}
o2_samw_mean = {}
sigma_samw = {}
for key, frame in data.iteritems():
    old_grid = retrieve_old_grid(frame)
    o2_regrid, temp, salt = regrid(frame, old_grid, new_grid)
    sigma_samw[key] = gsw.sigma0(salt, temp)
    
    o2_samw1 = ma.masked_where(sigma_samw[key] >= 27.0, o2_regrid )
    o2_samw_2d[key] = ma.masked_where(sigma_samw[key] <= 26.6, o2_samw1)
    o2_samw_mean[key] = np.nanmean(o2_samw_2d[key])
```

### Antarctic Bottom Water: 

```python
# New Grid 
lat = np.arange(-65.,-51.,1)
depth = np.arange(0,5000,100)   
XI_aabw, YI_aabw = np.meshgrid(lat,depth)
new_grid = (XI_aabw, YI_aabw)

o2_aabw_mean = {}
o2_aabw_2d = {}
sigma_aabw = {}
for key, frame in data.iteritems():
    old_grid = retrieve_old_grid(frame)
    o2_regrid, temp, salt = regrid(frame, old_grid, new_grid)
    sigma_aabw[key] = gsw.sigma0(salt, temp)
    
    o2_aabw = ma.masked_where(sigma_aabw[key] >= 27.6, o2_regrid )
    o2_aabw_2d[key] = ma.masked_where(sigma_aabw[key] <= 27.2, o2_aabw)
    o2_aabw_mean[key] = np.nanmean(o2_aabw_2d[key])
```

```python
clevs = np.arange(-20, 25, 5)

lats = np.arange(-65.,-51.,1)
depth = np.arange(0,5000,100)   

plt.figure(figsize=(14,6))
ax = plt.gca()
plt.contourf(lats, depth, ((o2_aabw_2d['I08_2005'] - o2_aabw_2d['I08_1994'])/o2_aabw_2d['I08_1994'])*100, clevs, cmap = 'RdBu_r', extend = 'both')
plt.colorbar()
CS = plt.contour(lats, depth, sigma_aabw['I08_1994'], colors = 'k', levels=[26.6, 27.0, 27.2, 27.6])
plt.clabel(CS, fontsize=9, inline=1)
plt.ylim([0, 2000])
ax.invert_yaxis()
plt.xlim([-65, -15])

lats = np.arange(-50.,-21.,1)
depth = np.arange(0,1500,100)   

plt.contourf(lats, depth, ((o2_samw_2d['I08_2005'] - o2_samw_2d['I08_1994'])/o2_samw_2d['I08_1994'])*100, clevs,cmap = 'RdBu_r', extend = 'both')
CS = plt.contour(lats, depth, sigma_samw['I08_1994'], colors = 'k', levels=[26.6, 27.0, 27.2, 27.6])
plt.clabel(CS, fontsize=9, inline=1)
plt.xlim([-65, -15])
```
