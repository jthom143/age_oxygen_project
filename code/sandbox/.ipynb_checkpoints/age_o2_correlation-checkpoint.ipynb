###### Load Data

```python
PATH = '/RESEARCH/chapter3/data/newCO2_control_800/'
o2 = iris.load_cube(PATH+'o2.nc')
age = iris.load_cube(PATH+'residency_age_surface.nc')
u = iris.load_cube(PATH+'u.nc')
v = iris.load_cube(PATH+'v.nc')
```

###### Change name of age time coordinate 

```python
age.coord('Time').rename('time')
```

```python
print o2
print age
```

```python
ulon = u.coord('longitude')
vlon = v.coord('longitude')
print ulon
```

```python
u_clim = u.collapsed('time', iris.analysis.MEAN)
v_clim = v.collapsed('time', iris.analysis.MEAN)
velocity = (u_clim ** 2 + v_clim ** 2) ** 0.5
velocity.rename('velocity')
```

```python
x = u.coord('longitude').points
y = u.coord('latitude').points
```

```python
plt.figure(figsize=(10,6))
ax = plt.axes(projection=ccrs.PlateCarree())

# Get the coordinate reference system used by the data
#transform = ulon.coord_system.as_cartopy_projection()
ax.set_extent((-80, 20, 0, 90), crs=ccrs.PlateCarree())

# Plot the wind speed as a contour plot
qplt.contourf(velocity[5,:,:], 20, cmap = 'jet')

# Add arrows to show the wind vectors
plt.quiver(x, y, u_clim[5,:,:].data, v_clim[5,:,:].data, pivot='middle')
ax.coastlines()
```

# Create correlation map for age vs o2 for each depth level

```python
correlation = iris.analysis.stats.pearsonr(age, o2, corr_coords='time') 
```

### Correlation at Surface (k=5, ~60m)

```python
plt.figure(figsize=(12,6))
qplt.contourf(correlation[5,:,:], cmap = 'RdBu_r')
```

### Correlation at intermediate depth (k=10, ~140m)

```python
plt.figure(figsize=(12,6))
qplt.contourf(correlation[10,:,:], cmap = 'RdBu_r')
```

### Correlation at depth (k=20, ~2000m)

```python
plt.figure(figsize=(12,6))
qplt.contourf(correlation[20,:,:], cmap = 'RdBu_r')
```

###### Correlation as a function of depth

Not entirely sure that taking the correlation across 3 dimesnions (time, latitude, and longitude) is physically relevant. Take with a grain of salt. 

```python
age.coord('latitude').guess_bounds()
age.coord('longitude').guess_bounds()
o2.coord('latitude').guess_bounds()
o2.coord('longitude').guess_bounds()
```

```python
grid_areas = iris.analysis.cartography.area_weights(age)
correlation_depth = iris.analysis.stats.pearsonr(age, o2, corr_coords=['time', 'latitude', 'longitude'], weights = grid_areas)
```

```python
print correlation_depth
```

```python
qplt.plot(correlation_depth)
```

### Correlation as a function of Depth and Latitude

```python
correlation_lat = iris.analysis.stats.pearsonr(age, o2, corr_coords=['time', 'longitude'], weights = grid_areas)
```

```python
qplt.contourf(correlation_lat, cmap = 'RdBu')
```

```python
lon_constraint = iris.Constraint(longitude=lambda y: -33 < y <= -30)
age_longitude45W = age.extract(lon_constraint)
o2_longitude45W = o2.extract(lon_constraint)
print age_longitude45W
```

```python
correlation_lat = iris.analysis.stats.pearsonr(age_longitude45W, o2_longitude45W, corr_coords='time')
qplt.contourf(correlation_lat, cmap = 'RdBu')
```

### Focus on Gulf Stream Region

```python
# Create a Cartopy crs for plain and rotated lat-lon projections.
plain_crs = ccrs.PlateCarree()
rotated_crs = ccrs.RotatedPole(pole_longitude=120.0, pole_latitude=45.0)

#correlation.coord('latitude').guess_bounds()
#correlation.coord('longitude').guess_bounds()
plt.figure(figsize=(12,6))

lons = age.coord('longitude').points
lats = age.coord('latitude').points

clevs = np.arange(-1, 1.05, 0.05)
ax = plt.subplot(111, projection=rotated_crs)
im = ax.contourf(lons, lats, correlation[10,:,:].data, clevs, transform=ccrs.PlateCarree(), cmap = 'RdBu_r')


ax.set_extent((-80, -55, 20, 50), crs=plain_crs)
ax.coastlines(resolution='50m', color='black', lw = 1.5)
ax.gridlines(color='lightgrey', linestyle='-')
plt.colorbar(im)
plt.show() 
```

```python
lat_constraint = iris.Constraint(latitude=lambda y: 20 < y < 30)
lon_constraint = iris.Constraint(longitude=lambda x: -75 < x < -20)
correlation_regional = correlation[5,:,:].extract(lat_constraint)
correlation_regional = correlation_regional.extract(lon_constraint)

clevs = np.arange(-1, 1.1, 0.1)
# Create a Cartopy crs for plain and rotated lat-lon projections.
plain_crs = ccrs.PlateCarree()
rotated_crs = ccrs.RotatedPole(pole_longitude=120.0, pole_latitude=45.0)
plt.figure(figsize=(12,10))
ax = plt.axes(projection=rotated_crs)

# Get the coordinate reference system used by the data
#transform = ulon.coord_system.as_cartopy_projection()
ax.set_extent((-80, 10, 0, 90), crs=ccrs.PlateCarree())

# Plot the wind speed as a contour plot
qplt.contourf(correlation_regional, clevs, cmap = 'RdBu_r')
ax.coastlines(resolution='50m', color='black', lw = 1.5)
```

```python
lat_constraint = iris.Constraint(latitude=lambda y: 20 < y < 30)
lon_constraint = iris.Constraint(longitude=lambda x: -75 < x < -20)
age_regional = age[:,5,:,:].extract(lat_constraint)
age_regional = age_regional.extract(lon_constraint)

o2_regional = o2[:,5,:,:].extract(lat_constraint)
o2_regional = o2_regional.extract(lon_constraint)

grid_areas = iris.analysis.cartography.area_weights(age_regional)
age_regional_average = age_regional.collapsed(['latitude', 'longitude'], iris.analysis.MEAN, weights = grid_areas)

o2_regional.coord('latitude').guess_bounds()
o2_regional.coord('longitude').guess_bounds()
grid_areas = iris.analysis.cartography.area_weights(o2_regional)
o2_regional_average = o2_regional.collapsed(['latitude', 'longitude'], iris.analysis.MEAN, weights = grid_areas)
```
