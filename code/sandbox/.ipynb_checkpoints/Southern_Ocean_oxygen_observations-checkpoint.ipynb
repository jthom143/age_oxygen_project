### Load Data - Transect P16

###### 1991 Cruise (Cruise ID: 31WTTUNES_2)

```python
fname ='5_p16s_hy1.csv'
path = '/RESEARCH/chapter3/data/GO_SHIP/P16/31WTTUNES_2/'
data = pd.read_csv(path+fname, header = 5, na_values='-999.000')
```

```python
o2_data_P16_1991 = data[['DATE', 'LATITUDE', 'LONGITUDE', 'CTDPRS', 'OXYGEN', 'OXYGEN_FLAG_W', 'CTDTMP', 'CTDSAL']].copy()
o2_data_P16_1991.columns = ['date', 'latitude', 'longitude', 'press', 'oxygen', 'oxygen_flag', 'temp', 'salt']
```

```python
# Drop row with units
o2_data_P16_1991 = o2_data_P16_1991.drop(data.index[[0]])

# Drop data that is flagged
o2_data_P16_1991 = o2_data_P16_1991[o2_data_P16_1991.oxygen_flag!=5]
o2_data_P16_1991 = o2_data_P16_1991[o2_data_P16_1991.oxygen_flag!=9]

# Remove data points that do not fall on line P16
o2_data_P16_1991 = o2_data_P16_1991[o2_data_P16_1991.longitude<=-148]
o2_data_P16_1991 = o2_data_P16_1991[o2_data_P16_1991.longitude>=-156]
```

###### 1992 Cruise (Cruise ID: 316N138_9)

```python
fname ='8_p16a_hy1.csv'
path = '/RESEARCH/chapter3/data/GO_SHIP/P16/316N138_9/'
data = pd.read_csv(path+fname, header = 5, na_values='-999.000')
```

```python
o2_data_P16_1992 = data[['DATE', 'LATITUDE', 'LONGITUDE', 'CTDPRS', 'OXYGEN', 'OXYGEN_FLAG_W', 'CTDTMP', 'CTDSAL']].copy()
o2_data_P16_1992.columns = ['date', 'latitude', 'longitude', 'press', 'oxygen', 'oxygen_flag', 'temp', 'salt']
```

```python
# Drop row with units
o2_data_P16_1992 = o2_data_P16_1992.drop(data.index[[0]])

# Drop data that is flagged
o2_data_P16_1992 = o2_data_P16_1992[o2_data_P16_1992.oxygen_flag!=5]

# Remove data points that do not fall on line P16
o2_data_P16_1992 = o2_data_P16_1992[o2_data_P16_1992.longitude<=-148]
o2_data_P16_1992 = o2_data_P16_1992[o2_data_P16_1992.longitude>=-156]
```

###### Combine the two sections to create one transect for the 1990s

```python
frames = [o2_data_P16_1991, o2_data_P16_1992]
o2_data_P16_1990s = pd.concat(frames)
```

### Regrid Each Hydrography Track onto Regular Grid (1 deg lat x 250 m)

```python
# Old Depth and Latitude Coordinates
depi = o2_data_P16_1990s.press.values
lati = o2_data_P16_1990s.latitude.values

# New Grid 
lat = np.arange(-72.,-16.,1)
depth = np.arange(0,5000,250)

XI, YI = np.meshgrid(lat,depth)

# Regrid
o2_data_P16_1990s_grid   = griddata((lati.flatten(), depi.flatten()), o2_data_P16_1990s.oxygen.values.flatten(), (XI,YI), 
                                    method='linear')
temp_data_P16_1990s_grid = griddata((lati.flatten(), depi.flatten()), o2_data_P16_1990s.temp.values.flatten(), (XI,YI), 
                                    method='linear')
salt_data_P16_1990s_grid = griddata((lati.flatten(), depi.flatten()), o2_data_P16_1990s.salt.values.flatten(), (XI,YI), 
                                    method='linear')
```

```python
sigma0_1990s = gsw.sigma0(salt_data_P16_1990s_grid, temp_data_P16_1990s_grid)

plt.figure(figsize=(14,6))
ax = plt.gca()
clevs = np.arange(100,300,20)
plt.contourf(XI, YI, o2_data_P16_1990s_grid , clevs,cmap = cmaps.viridis, extend = 'max')
cb = plt.colorbar()
CS = plt.contour(XI, YI, sigma0_1990s, colors = 'k', levels=[26.6, 27.0, 27.2, 27.6])
plt.clabel(CS, fontsize=9, inline=1)
plt.ylim([200, 1700])
plt.xlim([-65, -15])
ax.invert_yaxis()
plt.ylabel('Depth [dbars]', fontsize = 14)
cb.set_label('Oxygen [umol/kg]', fontsize = 14)
plt.xlabel('Latitude', fontsize = 14)
plt.title('P16S 1990s Observations', fontsize = 16)

sigma0_2005 = gsw.sigma0(salt_data_P16_2005_grid, temp_data_P16_2005_grid)

plt.figure(figsize=(14,6))
ax = plt.gca()
clevs = np.arange(100,300,20)
plt.contourf(XI, YI, o2_data_P16_2005_grid , clevs,cmap = cmaps.viridis, extend = 'both')
cb = plt.colorbar()
CS = plt.contour(XI, YI, sigma0_2005, colors = 'k', levels=[26.6, 27.0, 27.2, 27.6])
plt.clabel(CS, fontsize=9, inline=1)
plt.ylim([200, 1700])
plt.xlim([-65, -15])
ax.invert_yaxis()
plt.ylabel('Depth [dbars]', fontsize = 14)
cb.set_label('Oxygen [umol/kg]', fontsize = 14)
plt.xlabel('Latitude', fontsize = 14)
plt.title('P16S 2005 Observations', fontsize = 16)


sigma0_2015 = gsw.sigma0(salt_data_P16_2015_grid, temp_data_P16_2015_grid)

plt.figure(figsize=(14,6))
ax = plt.gca()
clevs = np.arange(100,300,20)
plt.contourf(XI, YI, o2_data_P16_2015_grid , clevs,cmap = cmaps.viridis, extend = 'both')
cb = plt.colorbar()
CS = plt.contour(XI, YI, sigma0_2015, colors = 'k', levels=[26.6, 27.0, 27.2, 27.6])
plt.clabel(CS, fontsize=9, inline=1)
plt.ylim([200, 1700])
plt.xlim([-65, -15])
ax.invert_yaxis()
plt.ylabel('Depth [dbars]', fontsize = 14)
cb.set_label('Oxygen [umol/kg]', fontsize = 14)
plt.xlabel('Latitude', fontsize = 14)
plt.title('P16S 2015 Observations', fontsize = 16)
```

# Calculate Precent Change in Each Watermass

### Subantarcitc Mode Water

###### 1990s:

```python
import numpy.ma as ma
# Old Depth and Latitude Coordinates
depi = o2_data_P16_1990s.press.values
lati = o2_data_P16_1990s.latitude.values

# New Grid 
lat = np.arange(-50.,-16.,1)
depth = np.arange(0,5000,250)

XI, YI = np.meshgrid(lat,depth)

# Regrid
o2_data_P16_1990s_grid_samw   = griddata((lati.flatten(), depi.flatten()), o2_data_P16_1990s.oxygen.values.flatten(), (XI,YI), 
                                    method='linear')
temp_data_P16_1990s_grid_samw  = griddata((lati.flatten(), depi.flatten()), o2_data_P16_1990s.temp.values.flatten(), (XI,YI), 
                                    method='linear')
salt_data_P16_1990s_grid_samw  = griddata((lati.flatten(), depi.flatten()), o2_data_P16_1990s.salt.values.flatten(), (XI,YI), 
                                    method='linear')

sigma0_1990s = gsw.sigma0(salt_data_P16_1990s_grid_samw, temp_data_P16_1990s_grid_samw)



o2_samw = ma.masked_where(sigma0_1990s >= 27.0, o2_data_P16_1990s_grid_samw )
o2_samw = ma.masked_where(sigma0_1990s <= 26.6, o2_samw)
o2_samw_1990 = np.nanmean(o2_samw)
```
