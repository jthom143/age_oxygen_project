### Define Functions:

```python
def load_data(fname, path, header_no, variables, new_names, lon_lims):
    # Load Data
    data = pd.read_csv(path+fname, header = header_no, na_values='-999.000')
    
    # Isolate Desired Variables
    var = data[variables].copy()
    
    # Rename variables if desired
    var.columns = new_names
    
    # Drop data that is flagged
    var = var[var.oxygen_flag!=5]
    var = var[var.oxygen_flag!=9]
    
    # Drop row with units
    var = var.drop(data.index[[0]])


    var = var[var.longitude<=lon_lims[0]]
    var = var[var.longitude>=lon_lims[1]] 
        
    return var
```

```python
def retrieve_old_grid(frame):
    depi = frame.press.values
    lati = frame.latitude.values
    old_grid = (lati.flatten(), depi.flatten())
    
    return old_grid
```

```python
def regrid(frame, old_grid, new_grid):
    o2_grid   = griddata(old_grid, frame.oxygen.values.flatten(), new_grid, 
                          method='linear')
    
    aou_grid   = griddata(old_grid, frame.aou.values.flatten(), new_grid, 
                           method='linear')
    
    sigma_grid   = griddata(old_grid, frame.sigma.values.flatten(), new_grid, 
                           method='linear')
    return o2_grid, aou_grid, sigma_grid
```

### Load Data:

```python
data_dict = {
             'P18_1994': {'fname': '0_31DSCG94_1_hy1.csv',
                          'path': '/RESEARCH/chapter3/data/GO_SHIP/P18/31DSCG94_1/', 'head_no':7},
             'P18_2007': {'fname': '6_33RO20071215_hy1.csv',
                          'path': '/RESEARCH/chapter3/data/GO_SHIP/P18/33RO20071215/', 'head_no':53}
            
             }
```

```python
data = {}
for key, info in data_dict.iteritems():  
    data[key] = load_data(info['fname'], info['path'], info['head_no'],
                          ['DATE', 'LATITUDE', 'LONGITUDE', 'CTDPRS', 'OXYGEN', 'OXYGEN_FLAG_W', 'CTDTMP', 'CTDSAL'], 
                          new_names=['date', 'latitude', 'longitude', 'press', 'oxygen', 'oxygen_flag', 'temp', 'salt'], 
                          lon_lims=[-100, -115])
```

### Regrid:

```python
# New Grid 
lat = np.arange(-72.,-16.,1)
depth = np.arange(0,5000,100)   
XI, YI = np.meshgrid(lat,depth)
new_grid = (XI, YI)
```

```python
o2_regrid = {}
sigma = {}
o2_aou = {}
for key, frame in data.iteritems():
    frame['oxygen'] = frame['oxygen'].astype(dtype=float)
    frame['temp'] = frame['temp'].astype(dtype=float)
    frame['salt'] = frame['salt'].astype(dtype=float)

    # Calculate Sigma
    frame['sigma'] = gsw.sigma0(frame.salt, frame.temp)
    
    # Calculate AOU
    frame['o2_sat'] = o2sat(frame.salt, frame.temp)
    frame['aou'] = frame.o2_sat - frame.oxygen
    
    
    old_grid = retrieve_old_grid(frame)
    o2_regrid[key], o2_aou[key], sigma[key] = regrid(frame, old_grid, new_grid)
```

### Oxygen Concentration:

```python
clevs = np.arange(100,300,20)

fig = plt.figure(figsize=(18,10))
ax1 = plt.subplot(2, 1, 1)
ax1.contourf(XI, YI, o2_regrid['P18_1994'], clevs, cmap = cmaps.viridis, extend = 'both')
plt.plot(data['P18_1994'].latitude, data['P18_1994'].press, '.', ls = 'none', color = 'grey')
CS = plt.contour(XI, YI, sigma['P18_1994'], colors = 'k', levels=[26.6, 27.0, 27.2, 27.6])
plt.clabel(CS, fontsize=9, inline=1)
plt.ylim([200, 1700])
plt.xlim([-65, -15])
ax1.invert_yaxis()
ax1.set_title('1994', fontsize = 16)


ax2 = plt.subplot(2, 1, 2)
im = ax2.contourf(XI, YI, o2_regrid['P18_2007'], clevs, cmap = cmaps.viridis, extend = 'both')
plt.plot(data['P18_2007'].latitude, data['P18_2007'].press, '.', ls = 'none', color = 'grey')
CS = plt.contour(XI, YI, sigma['P18_2007'], colors = 'k', levels=[26.6, 27.0, 27.2, 27.6])
plt.clabel(CS, fontsize=9, inline=1)
plt.ylim([200, 1700])
plt.xlim([-65, -15])
ax2.invert_yaxis()
ax2.set_title('2007', fontsize = 16)


cbaxes = fig.add_axes([0.94, 0.2, 0.02, 0.6]) 
cb = plt.colorbar(im, cax = cbaxes)
cb.set_label('Oxygen [umol/kg]', fontsize = 16)
```
