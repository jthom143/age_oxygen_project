### Calculate Age and Oxygen for CDW and SAMW

Define CDW and SAMW water masses following (approximately) same method used in Waugh et al., 2013: 

- SAMW: Water that lies on potential density surfaces between 26.6 - 27.0 and between latitudes 20S - 50S 
- CDW:  Water that lies on potential density surfaces between 27.2 - 27.7 and South of 50S. 

###### Regrid onto a smaller vertical grid

```python
def SAMW_smaller_grid(var_cube, z_grid, x_grid):
    
    # Dimensions of original cube
    time = var_cube.coord('time').points
    lons = var_cube.coord('longitude').points
    lats = var_cube.coord('latitude').points
    depths = var_cube.coord('tcell pstar').points
    
    new_grid = np.ones((len(time), len(z_grid), len(x_grid), len(lons)))
    
    for t in range(0, len(time)):
        for y in range(0,len(lons)):
            old_grid_data = var_cube[y,:,:,y].data

            #create mesh
            X, Y = np.meshgrid(lats, depths)
            XI, YI = np.meshgrid(x_grid, z_grid)

            #interp
            new = griddata((X.flatten(),Y.flatten()), old_grid_data.flatten(), (XI,YI),method='nearest')

            new_grid[t,:,:,y] = new
            
    return new_grid
```

```python
x_grid = np.arange(-50, -21, 1)
z_grid = np.arange(5, 2000, 50)

age_new_grid = SAMW_smaller_grid(age_SO, z_grid, x_grid)
o2_new_grid = SAMW_smaller_grid(o2_SO, z_grid, x_grid)
sigma0_new_grid = SAMW_smaller_grid(sigma0_SO, z_grid, x_grid)
```

###### Mask grid points outside of density threshold

```python
lat_constraint = iris.Constraint(latitude=lambda y: -50 < y <= -20)
age_samw_region = age.extract(lat_constraint)
o2_samw_region  = o2.extract(lat_constraint)
sigma0_samw_region = sigma0.extract(lat_constraint)

age_samw = ma.masked_where(sigma0_samw_region.data >= 27.0, age_samw_region.data)
age_samw = ma.masked_where(sigma0_samw_region.data <= 26.6, age_samw)

o2_samw = ma.masked_where(sigma0_samw_region.data >= 27.0, o2_samw_region.data)
o2_samw = ma.masked_where(sigma0_samw_region.data <= 26.6, o2_samw)
```

```python
slope, intercept, r_value, p_value, std_err = stats.mstats.linregress(age_samw.flatten(),o2_samw.flatten()*1e5)
print slope
print intercept
print r_value**2
```

```python
plt.scatter(age_samw, o2_samw*1e5)
plt.ylabel('Oxygen [umol/kg]', fontsize = 14)
plt.xlabel('Age [years]', fontsize = 14)
plt.title('SAMW Age- Oxygen Relationship', fontsize = 14)
x = np.arange(-100, 701, 1)
plt.plot(x, (slope*x)+intercept, color = 'k', lw = 2)
plt.xlim([-100, 700])
```

```python
lat_constraint = iris.Constraint(latitude=lambda y: y <= -50)
age_cdw_region = age.extract(lat_constraint)
o2_cdw_region  = o2.extract(lat_constraint)
sigma0_cdw_region = sigma0.extract(lat_constraint)

age_cdw = ma.masked_where(sigma0_cdw_region.data >= 27.7, age_cdw_region.data)
age_cdw = ma.masked_where(sigma0_cdw_region.data <= 27.2, age_cdw)

o2_cdw = ma.masked_where(sigma0_cdw_region.data >= 27.7, o2_cdw_region.data)
o2_cdw = ma.masked_where(sigma0_cdw_region.data <= 27.2, o2_cdw)
```

```python
slope, intercept, r_value, p_value, std_err = stats.mstats.linregress(age_cdw.flatten(),o2_cdw.flatten()*1e5)
print slope
print intercept
print r_value**2
```

```python
plt.scatter(age_cdw, o2_cdw*1e5)
plt.ylabel('Oxygen [umol/kg]', fontsize = 14)
plt.xlabel('Age [years]', fontsize = 14)
plt.title('CDW Age - Oxygen Relationship', fontsize = 14)
x = np.arange(-100, 801, 1)
plt.plot(x, (slope*x)+intercept, color = 'k', lw = 2)
plt.xlim([-100, 800])
```

### Basin by basin Analysis

Break the Southern Ocean down into individual basins and plot the age-oxygen relationship.

- Atlantic: 60W - 30E
- Indian: 30E - 130E
- Pacific: 130E - 60W

### Subantarctic Mode Water

###### Atlantic Ocean:

```python
lat_constraint = iris.Constraint(latitude=lambda y: -50 < y <= -20)
lon_constraint = iris.Constraint(longitude=lambda y: -60 < y <= 30)

age_Atl_samw_region = age.extract(lat_constraint)
age_Atl_samw_region = age_Atl_samw_region.extract(lon_constraint)

o2_Atl_samw_region  = o2.extract(lat_constraint)
o2_Atl_samw_region = o2_Atl_samw_region.extract(lon_constraint)

sigma0_Atl_samw_region = sigma0.extract(lat_constraint)
sigma0_Atl_samw_region = sigma0_Atl_samw_region.extract(lon_constraint)
```

```python
age_Atl_samw = ma.masked_where(sigma0_Atl_samw_region.data >= 27.0, age_Atl_samw_region.data)
age_Atl_samw = ma.masked_where(sigma0_Atl_samw_region.data <= 26.6, age_Atl_samw)

o2_Atl_samw = ma.masked_where(sigma0_Atl_samw_region.data >= 27.0, o2_Atl_samw_region.data)
o2_Atl_samw = ma.masked_where(sigma0_Atl_samw_region.data <= 26.6, o2_Atl_samw)
```

###### Indian Ocean

Indian Ocean is difficult because of longitude coordinate. Create two regions (one from positive longitude one from negative) and combine together

```python
# First Region
lat_constraint = iris.Constraint(latitude=lambda y: -50 < y <= -20)
lon_constraint = iris.Constraint(longitude=lambda y: 30 < y <= 76)

age_IO_samw_region1 = age.extract(lat_constraint)
age_IO_samw_region1 = age_IO_samw_region1.extract(lon_constraint)

o2_IO_samw_region1  = o2.extract(lat_constraint)
o2_IO_samw_region1 = o2_IO_samw_region1.extract(lon_constraint)

sigma0_IO_samw_region1 = sigma0.extract(lat_constraint)
sigma0_IO_samw_region1 = sigma0_IO_samw_region1.extract(lon_constraint)

# Second Region
lat_constraint = iris.Constraint(latitude=lambda y: -50 < y <= -20)
lon_constraint = iris.Constraint(longitude=lambda y: -285 < y <= -230)

age_IO_samw_region2 = age.extract(lat_constraint)
age_IO_samw_region2 = age_IO_samw_region2.extract(lon_constraint)

o2_IO_samw_region2  = o2.extract(lat_constraint)
o2_IO_samw_region2 = o2_IO_samw_region2.extract(lon_constraint)

sigma0_IO_samw_region2 = sigma0.extract(lat_constraint)
sigma0_IO_samw_region2 = sigma0_IO_samw_region2.extract(lon_constraint)


# Concatenate cubes together
cube_list = iris.cube.CubeList([age_IO_samw_region1, age_IO_samw_region2])
age_IO_samw_region = cube_list.concatenate_cube()

cube_list = iris.cube.CubeList([o2_IO_samw_region1, o2_IO_samw_region2])
o2_IO_samw_region = cube_list.concatenate_cube()

cube_list = iris.cube.CubeList([sigma0_IO_samw_region1, sigma0_IO_samw_region2])
sigma0_IO_samw_region = cube_list.concatenate_cube()
```

```python
age_IO_samw = ma.masked_where(sigma0_IO_samw_region.data >= 27.0, age_IO_samw_region.data)
age_IO_samw = ma.masked_where(sigma0_IO_samw_region.data <= 26.6, age_IO_samw)

o2_IO_samw = ma.masked_where(sigma0_IO_samw_region.data >= 27.0, o2_IO_samw_region.data)
o2_IO_samw = ma.masked_where(sigma0_IO_samw_region.data <= 26.6, o2_IO_samw)
```

###### Pacific Ocean

```python
lat_constraint = iris.Constraint(latitude=lambda y: -50 < y <= -20)
lon_constraint = iris.Constraint(longitude=lambda y: -230 < y <= -60)

age_Pac_samw_region = age.extract(lat_constraint)
age_Pac_samw_region = age_Pac_samw_region.extract(lon_constraint)

o2_Pac_samw_region  = o2.extract(lat_constraint)
o2_Pac_samw_region = o2_Pac_samw_region.extract(lon_constraint)

sigma0_Pac_samw_region = sigma0.extract(lat_constraint)
sigma0_Pac_samw_region = sigma0_Pac_samw_region.extract(lon_constraint)
```

```python
age_Pac_samw = ma.masked_where(sigma0_Pac_samw_region.data >= 27.0, age_Pac_samw_region.data)
age_Pac_samw = ma.masked_where(sigma0_Pac_samw_region.data <= 26.6, age_Pac_samw)

o2_Pac_samw = ma.masked_where(sigma0_Pac_samw_region.data >= 27.0, o2_Pac_samw_region.data)
o2_Pac_samw = ma.masked_where(sigma0_Pac_samw_region.data <= 26.6, o2_Pac_samw)
```

```python
plt.scatter(age_Pac_samw, o2_Pac_samw*1e5, color = 'g')
plt.scatter(age_Atl_samw, o2_Atl_samw*1e5, color = 'b')
plt.scatter(age_IO_samw, o2_IO_samw*1e5, color = 'r')
plt.ylabel('Oxygen [umol/kg]', fontsize = 14)
plt.xlabel('Age [years]', fontsize = 14)
plt.title('SAMW Age- Oxygen Relationship', fontsize = 14)
```
