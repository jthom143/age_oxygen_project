```python
depth = np.ones((len(lats), 25))*np.nan
for i in range(len(lats)):
    lat = lats[i]
    a = linew[linew.latitude==lat].depth.values
    depth[i,:len(a)] = a

oxygen = np.ones((len(lats), 25))*np.nan
for i in range(len(lats)):
    lat = lats[i]
    a = linew[linew.latitude==lat].oxygen.values
    oxygen[i,:len(a)] = a


# Interpolate
from scipy.interpolate import griddata

grid_x = np.array([5.07150000e+00,   1.52145000e+01,   2.53575000e+01,   3.55005000e+01,
   4.56435000e+01,   5.57865000e+01,   6.59295000e+01,   7.60725000e+01,
   8.76254925e+01,   1.09001150e+02,   1.42768544e+02,   1.96844110e+02,
   2.73066442e+02,   3.78403639e+02,   5.13549178e+02,   6.84154884e+02,
   8.89456286e+02,   1.13353807e+03,   1.41399425e+03,   1.73323094e+03,
   2.08716320e+03,   2.47655559e+03,   2.89575642e+03,   3.34407207e+03,
   3.81453442e+03,   4.30530531e+03,   4.80846744e+03,   5.32145264e+03])


#grid_x = np.mgrid[0:5000:100j]

oxygen_new = np.ones((len(grid_x),len(lats)))*np.nan
for i in range(len(lats)):
    lat = lats[i]
    depth = linew[linew.latitude==lat].depth.values

    b = griddata(depth, oxygen[i,:], grid_x, method='linear')
    oxygen_new[:,i] = b
    
x = np.ones((len(grid_x),1))*40

plt.figure(figsize=(14,8))
clevs = np.arange(100, 300.5, 0.5)
ax = plt.gca()
plt.contourf(np.array(lats), grid_x, oxygen_new, clevs, cmap = cmaps.viridis, extend = 'both')
ax.invert_yaxis()
cb = plt.colorbar()
cb.set_label('umol/kg', fontsize = 14)
plt.plot(x, grid_x, ls = '', marker = '*', color = 'k')
plt.plot(linew.latitude, linew.depth, ls = '', marker = '*', color = 'grey')
plt.ylabel('Depth [m]', fontsize = 14)
plt.xlabel('Latitude', fontsize = 14)
plt.title('GLODAPv2 Line W - Oxygen', fontsize = 14)
```

### Load Model Data

```python
import iris
PATH = '/RESEARCH/chapter3/data/newCO2_control_800/'
o2 = iris.load_cube(PATH+'o2.nc')
```

```python
model_lon = o2.coord('longitude')
model_lat = o2.coord('latitude')
```

```python
plt.figure(figsize=(12,10))
ax = plt.axes(projection=ccrs.PlateCarree())

# Get the coordinate reference system used by the data
#transform = ulon.coord_system.as_cartopy_projection()
ax.set_extent((-75, -65, 35, 45), crs=ccrs.PlateCarree())

# Plot the wind speed as a contour plot
qplt.contourf(o2[0,5,:,:], 20, cmap = 'jet')
plt.plot(linew_lons, linew_lats, marker = '*', transform=ccrs.PlateCarree())

ax.coastlines()
```

```python
def find_index(x,y):
    xi=np.searchsorted(model_lat.points,x)
    yi=np.searchsorted(model_lon.points,y)
    return xi,yi

lat_index = np.ones((len(linew_lats[:-6]), 1))*np.nan
lon_index = np.ones((len(linew_lats[:-6]), 1))*np.nan
for i in range(len(linew_lats[:-6])):
    thisLat, thisLon = find_index(linew_lats[i],linew_lons[i])
    lat_index[i] = thisLat
    lon_index[i] = thisLon
lat_index = lat_index.astype(int)
lon_index = lon_index.astype(int)
```

```python
o2_climatology = o2.collapsed('time', iris.analysis.MEAN)
```

```python
o2_new = np.ones((len(o2.coord('tcell pstar').points), len(lat_index)))* np.nan
for k in range(len(o2.coord('tcell pstar').points)):
    for i in range(len(lat_index)):
        o2_new[k,i] = o2_climatology[k, lat_index[i], lon_index[i]].data
```

```python

plt.figure(figsize=(14,8))
clevs = np.arange(100, 300.5, 0.5)
ax = plt.gca()
plt.contourf(linew_lats[:-6], o2.coord('tcell pstar').points, o2_new*1e6, clevs, cmap = cmaps.viridis, extend = 'both')
ax.invert_yaxis()
cb = plt.colorbar()
cb.set_label('umol/kg', fontsize = 14)
plt.ylabel('Depth [m]', fontsize = 14)
plt.xlabel('Latitude', fontsize = 14)
plt.title('Model Line W - Oxygen', fontsize = 14)
```
