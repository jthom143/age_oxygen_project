###### Load Packages

```python
import iris 
import numpy as np
import matplotlib.pyplot as plt
import iris.analysis.stats
import iris.quickplot as qplt
import seaborn as sns
import iris.analysis.cartography
import cartopy.crs as ccrs
from oceans import wrap_lon180
```

```python
import sys
sys.path.append('/RESEARCH/paper_ocean_heat_carbon/code/python')
import colormaps as cmaps
import utils
```

```python
%matplotlib inline
```

###### Load Data

```python
PATH = '/RESEARCH/chapter3/data/newCO2_control_800/'
o2 = iris.load_cube(PATH+'o2.nc')
age = iris.load_cube(PATH+'residency_age_surface.nc')
```

###### Change name of time coordinate

```python
age.coord('Time').rename('time')
```

###### Ocean Bling Data is too Long

```python
o2 = o2[-500:]
```

### Attempt to Isolate Atlantic Ocean

```python
basin_name = 'atlantic'
age_atlantic = utils.basin_mask(age, basin_name, is_basin=True, xp=None, yp=None)
o2_atlantic = utils.basin_mask(o2, basin_name, is_basin=True, xp=None, yp=None)
```

```python
lons = age.coord('longitude').points
lats = age.coord('latitude').points
ax = plt.axes(projection=ccrs.PlateCarree())
plt.contourf(lons, lats, age_atlantic[0,10,:,:].data, cmap = cmaps.viridis)
ax.coastlines()
plt.title('Ideal Age')
cb = plt.colorbar(orientation='horizontal')
cb.set_label('Years')
```

```python
ax = plt.axes(projection=ccrs.PlateCarree())
plt.contourf(lons, lats, o2_atlantic[0,10,:,:].data*1e3, cmap = cmaps.viridis)
ax.coastlines()
plt.title('Dissolved Oxygen')
cb = plt.colorbar(orientation='horizontal')
cb.set_label('mmols kg$^{-1}$')
```

### Zonally Average and Plot Latitude vs Depth Climatology, and Correlation

```python
age_zonal = age_atlantic.collapsed('longitude', iris.analysis.MEAN)
o2_zonal = o2_atlantic.collapsed('longitude', iris.analysis.MEAN)

o2_zonal_clim = o2_zonal.collapsed('time', iris.analysis.MEAN)
age_zonal_clim = age_zonal.collapsed('time', iris.analysis.MEAN)
```

```python
qplt.contourf(o2_zonal_clim, cmap = cmaps.viridis)
```

```python
qplt.contourf(age_zonal_clim, cmap = cmaps.viridis)
```

```python
correlation = iris.analysis.stats.pearsonr(age_zonal, o2_zonal, corr_coords='time') 
```

```python
clevs = np.arange(-1, 1.2, 0.2)
qplt.contourf(correlation, clevs, cmap = 'RdBu_r')
```
