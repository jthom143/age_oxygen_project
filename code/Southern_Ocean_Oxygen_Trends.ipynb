```python
# Load Packages and Such
import numpy as np
import pandas as pd
import sys
import gsw
import iris.quickplot as qplt
import matplotlib.pyplot as plt
from scipy.interpolate import griddata
sys.path.append('/RESEARCH/paper_ocean_heat_carbon/code/python')
import colormaps as cmaps
import numpy.ma as ma

sys.path.append('/RESEARCH/chapter3/functions')
from o2sat import o2sat


%matplotlib inline
```

```python
## Define Functions used in Analysis
def load_data(fname, path, header_no, variables, new_names, lon_lims):
    # Load Data
    data = pd.read_csv(path+fname, header = header_no, na_values='-999.000')
    
    # Isolate Desired Variables
    var = data[variables].copy()
    
    # Rename variables if desired
    var.columns = new_names
    
    # Drop data that is flagged
    var = var[var.oxygen_flag!=5]
    var = var[var.oxygen_flag!=9]
    
    # Drop row with units
    var = var.drop(data.index[[0]])


    var = var[var.longitude<=lon_lims[0]]
    var = var[var.longitude>=lon_lims[1]] 
        
    return var

def retrieve_old_grid(frame):
    depi = frame.press.values
    lati = frame.latitude.values
    old_grid = (lati.flatten(), depi.flatten())
    
    return old_grid

def regrid(frame, old_grid, new_grid):
    o2_grid   = griddata(old_grid, frame.oxygen.values.flatten(), new_grid, 
                          method='linear')
    
    temp_grid   = griddata(old_grid, frame.temp.values.flatten(), new_grid, 
                           method='linear')
    
    salt_grid   = griddata(old_grid, frame.salt.values.flatten(), new_grid, 
                           method='linear')
    return o2_grid, temp_grid, salt_grid
```

## Main Results: 

Calculated the change of both dissolved oxygen and apparent oxygen utilization (AOU) for each repeat hydrography transect in the Southern Ocean. Following figure shows the percent change per decade for both oxygen and AOU for two watermasses: Subantarctic Mode Water (SAMW) and Antarctic bottom Water (AABW). SAMW is defined as the water between latitudes 20S and 50S which lies between isopycnal surfaces 26.6 and 27.0. AABW is defined at the water South of 50S that lies between isopycnal surfaces 27.2 and 27.7. 

Apparent Oxygen Utilization is defined as the following: 
$$ AOU = O_2^{sat} - O_2 $$ 

where $O_2$ is the measured dissolved oxygen and $O_2^{sat}$ is the calculated oxygen saturation. The oxygen saturation is calculated from the measured salinity and potential temperature following the emperical formulation described in Weiss, 1970. 

For both dissolved oxygen and AOU, little change in concentration is seen in the AABW water mass. SAMW on the other hand shows general decreases in dissolved oxygen along all lines and increases in AOU along all lines. The percentage increases in AOU are significantly larger than those seen in dissolved oxygen.

```python
# Load Data
data = {}
for key, info in data_dict.iteritems():  
    if key == 'P16_1991' or key =='P16_1992' or key =='P16_2005' or key =='P16_2015':
        YY = [-148, -156]
    elif key == 'P18_1994' or key =='P18_2007' or key=='P18_2016':
        YY = [-100, -105]
    elif key == 'A16_1989' or key =='A16_2005' or key=='A16_2014':
        YY = [-22, -37]
    elif key == 'I08_1994' or key =='I08_2005' or key=='I08_2016':
        YY = [100, 80]
    
    data[key] = load_data(info['fname'], info['path'], info['head_no'],
                          ['DATE', 'LATITUDE', 'LONGITUDE', 'CTDPRS', 'OXYGEN', 'OXYGEN_FLAG_W', 'CTDTMP', 'CTDSAL'], 
                          new_names=['date', 'latitude', 'longitude', 'press', 'oxygen', 'oxygen_flag', 'temp', 'salt'], 
                          lon_lims=YY)

# Join P16 1990s data    
data.update({'P16_1990': pd.concat([data['P16_1991'], data['P16_1992']])})
del data['P16_1991']
del data['P16_1992']
```

```python
### Water Mass Analysis: 

# SAMW: 
# New Grid 
lat = np.arange(-50.,-21.,1)
depth = np.arange(0,5000,250)   
XI, YI = np.meshgrid(lat,depth)
new_grid = (XI, YI) 

o2_samw_mean = {}
o2_aou_samw = {}
for key, frame in data.iteritems():
    old_grid = retrieve_old_grid(frame)
    o2_regrid, temp, salt = regrid(frame, old_grid, new_grid)
    sigma_samw = gsw.sigma0(salt, temp)
    
    o2_samw = ma.masked_where(sigma_samw >= 27.0, o2_regrid )
    o2_samw = ma.masked_where(sigma_samw <= 26.6, o2_samw)
    o2_samw_mean[key] = np.nanmean(o2_samw)
    
    # Calculate o2_saturation
    o2_sat = o2sat(salt, temp)
    
    # Calculate AOU
    o2_aou = o2_sat - o2_regrid
    
    o2_aou = ma.masked_where(sigma_samw >= 27.0, o2_aou )
    o2_aou = ma.masked_where(sigma_samw <= 26.6, o2_aou)
    
    o2_aou_samw[key] = np.nanmean(o2_aou)
    
    

# AABW:
# New Grid 
lat = np.arange(-65.,-51.,1)
depth = np.arange(0,5000,250)   
XI, YI = np.meshgrid(lat,depth)
new_grid = (XI, YI)

o2_aabw_mean = {}
o2_aou_aabw = {}
for key, frame in data.iteritems():
    old_grid = retrieve_old_grid(frame)
    o2_regrid, temp, salt = regrid(frame, old_grid, new_grid)
    sigma_aabw = gsw.sigma0(salt, temp)
    
    o2_aabw = ma.masked_where(sigma_aabw >= 27.7, o2_regrid )
    o2_aabw = ma.masked_where(sigma_aabw <= 27.2, o2_aabw)
    o2_aabw_mean[key] = np.nanmean(o2_aabw)
    
    # Calculate o2_saturation
    o2_sat = o2sat(salt, temp)
    
    # Calculate AOU
    o2_aou = o2_sat - o2_regrid

    o2_aou = ma.masked_where(sigma_aabw >= 27.7, o2_aou )
    o2_aou = ma.masked_where(sigma_aabw <= 27.2, o2_aou)
    
    o2_aou_aabw[key] = np.nanmean(o2_aou)
```
